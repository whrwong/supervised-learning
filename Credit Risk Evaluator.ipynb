```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
#print(train_df)
print(test_df)
```

```python
train_df.info()
```

```python
# Convert categorical data to numeric and separate target feature for training data
numeric_train = pd.get_dummies(train_df)
#columns=["gender"]
numeric_train.info()
#print(numeric_train)

# target feature = credit status
X_train = numeric_train["loan_status_low_risk"]
print(X_train)

Y_train = numeric_train[["loan_amnt", "int_rate", "hardship_flag_N", "hardship_flag_Y", "annual_inc", "pymnt_plan_n", "debt_settlement_flag_N", "debt_settlement_flag_Y"]]
print(Y_train)


```

```python
# Convert categorical data to numeric
numeric_test = pd.get_dummies(test_df)
numeric_test.info()
print(numeric_test)

#test_df_dummy.describe()
new_numeric_test = numeric_test.assign(debt_settlement_flag_Y=0)
new_numeric_test.info()

# separate target feature for testing data
X_test = new_numeric_test["loan_status_low_risk"]
print(X_test)

Y_test = new_numeric_test[["loan_amnt", "int_rate", "hardship_flag_N", "hardship_flag_Y", "annual_inc", "pymnt_plan_n", "debt_settlement_flag_N", "debt_settlement_flag_Y"]]
print(Y_test)
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(Y_train, X_train)
print(f"Training Data Score: {classifier.score(Y_train, X_train)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10, max_depth=None, min_samples_split=2, random_state=0)
scores = cross_val_score(clf, Y_train, X_train, cv=5)
print(scores.mean())
```

```python
# Scale the data
# normalizing means to create the same scale to compare data 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
print(scaler.fit(Y_train))
print(scaler.mean_)
scaled_Y_train = scaler.transform(Y_train)
print(scaled_Y_train)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression()
classifier.fit(scaled_Y_train, X_train)
print(f"Training Data Score: {classifier.score(scaled_Y_train, X_train)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10, max_depth=None, min_samples_split=2, random_state=0)
scores = cross_val_score(clf, scaled_Y_train, X_train, cv=5)
print(scores.mean())
```
